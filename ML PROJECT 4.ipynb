TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x=a.drop('Class',axis='columns')
x
```

```python
y=a.Class
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
xtrain,xtest,ytrain,ytest
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

SVC

```python
from sklearn.svm import SVC
```

```python
model2=SVC()
model2.fit(xtrain,ytrain)
```

```python
model2.predict(xtest)
```

```python
model2.score(xtest,ytest)
```

KERNALS(LINEAR,POLY,RBF)

LINEAR

```python
model3=SVC(kernel='linear')
model3.fit(xtrain,ytrain)
```

```python
model3.predict(xtest)
```

```python
model3.score(xtest,ytest)
```

CONFUSSION MATRIX OF LINEAR

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
```

```python
y_p=model3.predict(xtest)
y_p
```

```python
cm=confusion_matrix(y_p,ytest)
cm
```

```python
import seaborn as sn
```

```python
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
plt.figure(figsize=(20,7))
```

VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Count of Fraud vs Non-Fraud transactions

```python
print(a['Class'].value_counts())
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=a)
plt.title('Number of Fraud vs Non-Fraud Transactions')
plt.xticks([0,1], ['Non-Fraud (0)', 'Fraud (1)'])
plt.show()
```

Distribution of Transaction Amount by Class

```python
plt.figure(figsize=(8,5))
sns.histplot(data=a, x='Amount', hue='Class', bins=50, log_scale=True, element='step', stat='density')
plt.title('Transaction Amount Distribution (Log Scale) by Class')
plt.xlabel('Amount')
plt.show()
```

Time vs Amount colored by Class (scatter plot)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=a, alpha=0.5, palette=['blue','red'])
plt.title('Transaction Amount over Time by Class')
plt.show()
```

Correlation heatmap of features V1 to V10 (subset for clarity)

```python
plt.figure(figsize=(12,8))
sns.heatmap(a.loc[:, 'V1':'V10'].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Features V1 to V10')
plt.show()
```
